```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt

# JUPYTER NOTEBOOK ONLY
# %matplotlib inline
```

```python
x = [0.69,0.91,2.10,24,23.24]
y = [2.77,1.54,5.34,50.16,52.24]
```

```python
plt.plot(x,y)

# Axis and ticks
plt.xlim(0,60)
plt.ylim(0,60)


# Labeling
plt.title('Runs')
plt.xlabel('MAgP')
plt.ylabel('P[0.1]')
```

```python
plt.plot(x,y,color='red')

# Axis and ticks
plt.xlim(0,60)
plt.ylim(0,60)


# Labeling
plt.title('Runs')
plt.xlabel('MAgP')
plt.ylabel('P[0.1]');
```

```python
plt.plot(x,y,color='red',marker='o',markersize=20,linestyle='-')

# Axis and ticks
plt.xlim(0,60)
plt.ylim(0,60)


# Labeling
plt.title('Runs')
plt.xlabel('MAgP')
plt.ylabel('P[0.1]');
```
